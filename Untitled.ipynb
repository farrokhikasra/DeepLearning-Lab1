```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,10000)

y = np.exp(-pow(x-2, 2))+np.exp(-pow(x-6, 2)/10)+1/(pow(x, 2)+1)

fig = plt.figure()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Exponential function')
plt.plot(x,y, 'b')
plt.show()
```

```python
#Load sklearn data
from sklearn.datasets import load_wine
from sklearn.manifold import TSNE

data = load_wine()


print(data.target.shape)
print(data.data.shape)

print(type(data.data[0, 0]))

print(data.data)

X_embedded = TSNE(n_components=2, init='random').fit_transform(data.data) #I tried to add learning_rate='auto' but code was not workind

print(X_embedded.shape)



first = []
second = []
third = []

for i in range(len(data.target)):

    if data.target[i] == 0:
        first.append(list(X_embedded[i, :]))
    if data.target[i] == 1:
        second.append(list(X_embedded[i, :]))
    if data.target[i] == 2:
        third.append(list(X_embedded[i, :]))

plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter plot of embedded data')

plt.scatter(np.array(first)[:, 0], np.array(first)[:, 1])
plt.scatter(np.array(second)[:, 0], np.array(second)[:, 1])
plt.scatter(np.array(third)[:, 0], np.array(third)[:, 1])


plt.legend(["First Cluster", "Second Cluster", "Third Cluster"])
plt.show()
```
